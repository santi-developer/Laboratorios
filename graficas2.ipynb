```python
from matplotlib import pyplot as plt
import pandas as pd

#leer los datos y meterlos en un dataframe
df = pd.read_csv('datos.txt', delimiter='\t')
```

# Grafica de distribucion de notas con un boxplot 

```python
# Definir el orden deseado de las materias
orden_materias = ['Matemáticas', 'Historia', 'Ciencias', 'Lenguaje']

# Ordenar el DataFrame según el orden de las materias
df['materia'] = pd.Categorical(df['materia'], categories=orden_materias, ordered=True)
df = df.sort_values('materia')

# Crear un boxplot
plt.figure(figsize=(10, 6))  # Tamaño opcional del gráfico
df.boxplot(column='nota', by='materia')

# Personalizar el gráfico
plt.title('Distribución de Notas por Materia')
plt.xlabel('Materia')
plt.ylabel('Nota')

# Rotar las etiquetas del eje X para una mejor legibilidad
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```

# Grafica de distribucion de aprobados con un pie chart

```python
 # Calcular la cantidad de estudiantes aprobados y no aprobados
aprobados = (df['aprobado'] == 'Sí').sum()
no_aprobados = (df['aprobado'] == 'No').sum()

# Crear un pie chart
plt.figure(figsize=(6, 6))  # Tamaño opcional del gráfico
plt.pie([aprobados, no_aprobados], labels=['Aprobados', 'No Aprobados'], autopct='%1.1f%%', startangle=90)

# Personalizar el gráfico
plt.title('Distribución de Aprobados vs. No Aprobados')

# Mostrar el gráfico
plt.show()
```
